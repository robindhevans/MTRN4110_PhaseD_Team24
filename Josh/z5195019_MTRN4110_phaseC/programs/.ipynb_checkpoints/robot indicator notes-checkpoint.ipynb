```python
def robot_indicator(center, direction, maze_rgb):
    robot_char = ' '
    #circle
    omega = np.arange(0.0, 6.3, 0.02)
    for om in omega:
        for rad in range(30,35):
            xplot = center[1] + (rad * math.sin(om))
            yplot = center[0] + (rad * math.cos(om))
            maze_rgb[int(xplot),int(yplot)] = [240,21,244]
    
    
    # South
    if(direction == 0):
        robot_char = 'V'
        tip = center[1] + 15
        for thick in range(0,5):
            for x in range(0, 22):
                maze_rgb[tip - x + thick, center[0] - x] = [240,21,244]
            for x in range(0, 22):
                maze_rgb[tip - x + thick, center[0] + x] = [240,21,244]
    # North
    elif(direction == 2):
        robot_char = '>'
        tip = center[1] - 15
        for thick in range(0,5):
            for x in range(0, 22):
                maze_rgb[tip + x - thick, center[0] - x] = [240,21,244]
            for x in range(0, 22):
                maze_rgb[tip + x - thick, center[0] + x] = [240,21,244]
    
    # East
    elif(direction == 1):
        robot_char = '^'
        tip = center[0] + 15
        for thick in range(0,5):
            for x in range(0, 22):
                maze_rgb[center[1] - x, tip - x + thick] = [240,21,244]
            for x in range(0, 22):
                maze_rgb[center[1] + x, tip - x + thick] = [240,21,244]

    # West
    elif(direction == 3):
        robot_char = '<'
        tip = center[0] - 20
        for thick in range(-2,3):
            for rad in range(0,35):
                x = tip + (rad*math.cos(0.5))
                yup = center[1] + (rad*math.sin(0.5)) + thick
                ydwn = center[1] - (rad*math.sin(0.5)) + thick
                maze_rgb[int(yup), int(x)] = [240,21,244]
                maze_rgb[int(ydwn), int(x)] = [240,21,244]
        
        #for thick in range(0,5):
            #for x in range(0, 22):
                #maze_rgb[center[1] - x, tip + x - thick] = [240,21,244]
            #for x in range(0, 22):
                #maze_rgb[center[1] + x, tip + x - thick] = [240,21,244]
```
